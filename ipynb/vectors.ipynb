```python
import numpy as np
```

```python
#vectors in python


#column vectors
#row vectors

#scalars

class Value:
    def __init__(self, data):
        self.data = data

    def __repr__(self):
        return f"Value(data={self.data})"

    def __add__(self, other):
        sum = Value(self.data + other.data)
        return sum
    
    def __mul__(self, other):
        product = Value(self.data * other.data)
        return product
```

```python
#scalar = whole number
# scalar number its basically a whole number that scales a vector or a matrix for this case its 3 and 4
a = Value(3)
b = Value(4)
```

```python
#and for the vectors here it is
#here is an example of a row vector 
#so the expected shape for this one is (3)

row_va = np.array([1, 3, 4])
print(f"this is a row vector {row_va}")
print(f"the shapes of it is {row_va.shape}")

row_va = Value(np.array([1, 3, 4]))
row_vb = Value(np.array([1, 3, 4]))

row_va + row_vb
```

```python
#and for the column vectors here it is
#this is a quick example of a column vector
#so the expected shape for this one is (3 = row, 2 = col)

col_va = np.array([[1, 2], [2, 3], [3, 4]])
print(f"the shape for this vector is {col_va.shape}")

col_vb = np.copy(col_va) + 1 # ([[2, 3], [3, 4], [4, 5]])
col_va = Value(col_va)
col_vb = Value(col_vb)

col_vb + col_va
```

```python
#now transpose our vector to a column
row_va = np.array([[1, 3, 4]])
print(row_va)
print(row_va.shape)
row_va = row_va.T
print(row_va)
print(row_va.shape)
```

```python
simple_ra = Value(np.array([[1, 2]]))
simple_rb = Value(np.array([[1, 2]]))

print(f"the product is : {simple_ra * simple_rb}")
```
