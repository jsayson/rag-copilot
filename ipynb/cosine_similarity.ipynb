```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
class Value:
    def __init__(self, data):
        self.data = data
    
    def __repr__(self):
        return f"data: {self.data}"
    
    def __add__(self, other):
        return Value(self.data + other.data)
    
    def __sub__(self, other):
        return Value(self.data - other.data)
    
    def __mul__(self, other):
        return Value(self.data * other.data)
    
    def __truediv__(self, other):
        return Value(self.data / other.data)
    
    def __pow__(self, other):
        return Value(self.data ** other.data)
    
    def dot(self, other):
        assert len(self.data) == len(other.data)
        val = Value(0)
        for a, b in zip(self, other):
            val += a * b
        return val
    
    def sqrt(self):
        return Value(math.sqrt(self.data))
    
    def cosine(self, other):
        cos = Value.dot(self, other) / (Value.dot(self, self).sqrt() * Value.dot(other, other).sqrt())
        return cos
```

```python
# cosine similarity is a way of finding how similar something is for example: words; by computing the cosine of the angle between them
# in a table
# words like (i love magic) and (magic ball)
# since there are 4 words then we create a simple table for it 
# [i, love, magic, ball]
# first sentence its gonna be like np.array([[1, 1, 1, 0]])
# second sentence it is np.array([[0, 0, 1, 1]])
# to compute the cosine similirity  = ai * bi / squareroot of ai ^ 2  * squareroot bi ^ 2

f_v = np.array([[Value(1), Value(1), Value(1), Value(0)]]).flatten()
s_v = np.array([[Value(1), Value(1), Value(1), Value(0)]]).flatten()


print(f"{Value.cosine(f_v, s_v)}")
```
